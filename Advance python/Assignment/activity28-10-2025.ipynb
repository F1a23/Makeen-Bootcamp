6. Adding / Modifying Columns

```python
#Create a column total_marks which is the sum of all marks columns.
df["total_marks"] = df[["sql_marks", "excel_marks", "python_marks", "power_bi_marks", "english_marks"]].sum(axis=1)
df
```

```python
#Create a column average_marks which is the mean of all marks columns.
df["average_marks"] = df[["sql_marks", "excel_marks", "python_marks", "power_bi_marks", "english_marks"]].mean(axis=1)
df
```

```python
#Create a column passed which is Yes if average_marks >= 75 else No .
import numpy as np
df["passed"] = np.where(df["average_marks"] >= 75, "Yes", "No")
df
```

8. Aggregation

```python
#Calculate the average Python marks.
average_python = df["python_marks"].mean()
print("Average Python Marks:", average_python)
```

```python
#Find the maximum Excel marks
maximun_Excel = df["python_marks"].max()
print("Maximum Excel marks:", maximun_Excel)
```

```python
#Calculate the average total marks by location.
df.groupby('location')['total_marks'].mean()
```

```python
#Count how many students scored more than 90 in SQL.
count_sql = (df["sql_marks"] > 90).sum()

print("Number of students is:", count_sql)
```

10. Challenge: Combining Skills

```python
#Find the top 5 students with highest average marks.
top5 = df.sort_values(by="average_marks", ascending=False).head(5)
top5
```

```python
#Find students from Los Angeles who scored more than 80 in all subjects.

marks_cols = ["sql_marks", "excel_marks", "python_marks", "english_marks"]

la_students = df[(df["location"] == "Los Angeles") & (df[marks_cols] > 80).all(axis=1)]

la_students
```

```python
#Find students whose total_marks is above 450 and age <= 23, then sort by total_marks descending.
sorted_students = df.loc[
    (df["total_marks"] > 450) & (df["age"] <= 23)
].sort_values("total_marks", ascending=False)
sorted_students
```
