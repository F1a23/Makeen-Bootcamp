```python
import pymysql
import pandas as pd
from datetime import datetime

# -------------------------------
# Read data from Excel file
# -------------------------------
file_path = "weather_report.xlsx"   # ← اسم الملف
data = pd.read_excel(file_path)     # ← يقرأ من Excel مباشرة

# -------------------------------
# Connect to MySQL database
# -------------------------------
connection = pymysql.connect(
    host="localhost",
    user="root",
    password="Oman99690050#",
    database="pythondb"
)

cursor = connection.cursor()

# -------------------------------
# Create table if it doesn't exist
# -------------------------------
create_table_query = """
CREATE TABLE IF NOT EXISTS weather_reports (
    id INT AUTO_INCREMENT PRIMARY KEY,
    city VARCHAR(255),
    temperature_c FLOAT,
    feels_like_c FLOAT,
    humidity INT,
    description VARCHAR(255),
    wind_speed FLOAT,
    timestamp DATETIME
);
"""
cursor.execute(create_table_query)

# -------------------------------
# Insert Excel data into MySQL
# -------------------------------
insert_query = """
INSERT INTO weather_reports (
    city, temperature_c, feels_like_c, humidity, description, wind_speed, timestamp
)
VALUES (%s, %s, %s, %s, %s, %s, %s);
"""

for _, row in data.iterrows():
    cursor.execute(insert_query, (
        row["city"],
        row["temperature_c"],
        row["feels_like_c"],
        row["humidity"],
        row["description"],
        row["wind_speed"],
        row["timestamp"]
    ))

connection.commit()
connection.close()

print("Weather data read from Excel and stored in MySQL successfully.")
```
