```python
import pandas as pd
import numpy as np
```

```python
#Q1: Create Series
 #Create a pandas Series for each city with daily temperatures.
```

```python
days = ["Day1", "Day2", "Day3", "Day4", "Day5", "Day6", "Day7"]

muscat = pd.Series([35, 36, 34, 33, 37, 38, 36], index=days, name="Muscat")
sohar  = pd.Series([28, 29, 27, 30, 31, 28, 29], index=days, name="Sohar")
nizwa  = pd.Series([40, 42, 39, 41, 43, 40, 41], index=days, name="Nizwa")

# add all togther 
series_list = [muscat, sohar, nizwa]
series_list

print("City", end="\t") # to be in one line with days
# loop for days:
for d in days:
    print(d, end="\t") #\t for space
print()

for c in series_list:
    print(c.name,end="\t")  # name : muscat,sohar ,nizwa
    
    for value in c:
        print(value, end="\t")
    print()
```

```python
# Q2: Average Temperature
# Calculate the average temperature for each city using NumPy.
```

```python
#for average use mean
avg_muscat=float(round(np.mean(muscat)))
avg_sohar=float(round(np.mean(sohar)))
avg_nizwa=float(round(np.mean(nizwa)))
print("Muscat",avg_muscat)
print("Sohar",avg_sohar)
print("Nizwa",avg_nizwa)
```

```python
# Q3: Cities above 35°C average
# Identify which cities had an average temperature above 35°C.
```

```python
print("Cities with average temperature above 35°C:")
if avg_muscat > 35:
    print("Muscat")
if avg_sohar > 35:
    print("Sohar")
if avg_nizwa > 35:
    print("Nizwa")
```

```python
# Q4: Hottest day per city
# Find the day (index) with the maximum temperature for each city using NumPy.
```

```python
for c in series_list:
    max_temp = c.max()  
    hot_day = ""       

    for d in days:
        if c[d] == max_temp:
            hot_day = d
            break
    print(c.name, ":", hot_day, "=", max_temp, "°C")
```

```python
#Q5: Boolean Series above average
# Create a boolean Series for each city showing True for days above the cityʼs average temperature and False otherwise.
```

```python
for c in series_list:
    avg=c.mean()
    print("\n", c.name, "average =", avg)
    for d in days:
        if c[d] > avg:
            print(d, True)
        else:
            print(d, False)
```

```python
# Q6: Count above-average days
# Using the boolean Series from Q5, count how many days each city had above-average temperatures.
```

```python
for c in series_list:
    avg = c.mean()
    count = 0
    for d in days:
        if c[d] > avg:
            count += 1
    print(c.name + ":", count)
```

```python
# Q7: Count days within a range
 #Count how many days each city had temperatures between 35°C and 40°C (inclusive).
```

```python
for c in series_list:
    count = 0
    for d in days:
        if c[d] >= 35 and c[d] <= 40:
            count += 1
    print(c.name + ":", count)
```

```python
# Q8: Compare two cities
#Create a boolean Series that shows True for days when Muscat was hotter than Sohar.
#Count how many days this happened.
```

```python
count = 0
for d in days:
    if muscat[d] > sohar[d]:
        print(d, True)
        count += 1
    else:
        print(d, False)

print("Total days Muscat hotter than Sohar:", count)
```

```python
#Q9: Extreme hot days
# Create a boolean Series for each city that shows True for temperatures above 1.1×average temperature 10% above average).
# Count how many extreme days each city had.
```

```python
for c in series_list:
    avg = c.mean()           
    limit = 1.1 * avg         
    count = 0
    print("\n", c.name, "limit =", limit)
    for d in days:
        if c[d] > limit:
            print(d, True)
            count += 1
        else:
            print(d, False)
    print(c.name + ":", count)
```

```python
# Q10: Coolest day overall
 #Find the day with the lowest temperature across all cities.
 #Indicate which city had that lowest temperature
```

```python
lowest_temp = 999
cool_day = ""
cool_city = ""

for c in series_list:
    for d in days:
        if c[d] < lowest_temp:
            lowest_temp = c[d]
            cool_day = d
            cool_city = c.name

print(cool_day + ",", cool_city + ",", str(lowest_temp) + "°C")
```

```python
# Q11: Difference to average
# For each city, create a Series showing the difference between each dayʼs temperature and the cityʼs average temperature.
# Identify the day with the largest positive difference for each city.
```

```python
for c in series_list:
    avg = c.mean()
    print("\n", c.name, "average =", avg)
    diff = []        
    for d in days:
        difference = c[d] - avg
        diff.append(difference)
        print(d, ":", difference)
    
    max_diff = max(diff)
    max_day = days[diff.index(max_diff)]
    print("Largest positive difference:", max_day, "=", max_diff)
```
