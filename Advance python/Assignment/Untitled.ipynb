```python
import requests
import pandas as pd


API_KEY = "e2ff8e020abb574b3f2eb97c686b2ba6"  


city = input("Enter city name: ")

url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}&units=metric"


response = requests.get(url)


data = response.json()

if str(data.get("cod")) != "200":
    print("City not found or API key invalid.")
    print("Message from API:", data.get("message"))
else:

    city_name = data["name"]                      # city name returned from API
    temperature = data["main"]["temp"]           # current temperature
    feels_like = data["main"]["feels_like"]      # feels like temperature
    humidity = data["main"]["humidity"]          # humidity %
    description = data["weather"][0]["description"]  # weather description
    wind_speed = data["wind"]["speed"]           # wind speed


    print("\nWeather Report:")
    print("City:", city_name)
    print("Temperature:", temperature, "째C")
    print("Feels Like:", feels_like, "째C")
    print("Humidity:", humidity, "%")
    print("Weather:", description)
    print("Wind Speed:", wind_speed, "m/s")


    weather_row = {
        "City": city_name,
        "Temperature_C": temperature,
        "Feels_Like_C": feels_like,
        "Humidity_%": humidity,
        "Description": description,
        "Wind_Speed_m_s": wind_speed
    }

df = pd.DataFrame([weather_row])  # list of one dict => one-row DataFrame

df.to_excel("weather_report.xlsx", index=False)

print("\nData saved to 'weather_report.xlsx'")
```
