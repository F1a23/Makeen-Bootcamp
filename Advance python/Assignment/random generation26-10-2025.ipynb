```python
#استخدام المجموع التراكمي

n =int(input("Enter starting number: ")) #starting number

numbers = []  #empty list to add the numbers
for i in range(10):
    n = (n * 4 + 5) % 100  # %100 means all numbers stays between 0 and 99.
    numbers.append(n)        

print("Generated numbers:", numbers)
```
