```python
import kagglehub
import tensorflow as tf
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# =========================================================
# Q1) Download the dataset and print the saved path
# =========================================================
path = kagglehub.dataset_download("paultimothymooney/chest-xray-pneumonia")
print("Path to dataset files:", path)
DATA_DIR = path + "/chest_xray"
```

```python
# =========================================================
# Q2) Read/Load the dataset (Train / Validation / Test)
# =========================================================
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
SEED = 42

train_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR + "/train",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode="binary",   # NORMAL=0 , PNEUMONIA=1
    shuffle=True,
    seed=SEED
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR + "/val",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode="binary",
    shuffle=True,
    seed=SEED
)

test_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR + "/test",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode="binary",
    shuffle=False
)

print("Classes:", train_ds.class_names)  # ['NORMAL', 'PNEUMONIA']

# Improve speed
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.cache().shuffle(1000).prefetch(AUTOTUNE)
val_ds   = val_ds.cache().prefetch(AUTOTUNE)
test_ds  = test_ds.cache().prefetch(AUTOTUNE)
```

```python
# =========================================================
# Q3) Data Exploration: Display images from each class
# =========================================================
images, labels = next(iter(train_ds))

plt.figure(figsize=(9,4))
for i in range(6):
    ax = plt.subplot(2,3,i+1)
    ax.imshow(images[i].numpy().astype("uint8"))
    ax.set_title("NORMAL" if labels[i].numpy()==0 else "PNEUMONIA")
    ax.axis("off")
plt.tight_layout()
plt.show()
```
