```python
import os                                  # Work with file paths
import glob                                # Search for files recursively
import time                                # Measure training time (bonus)
import numpy as np                         # Arrays and math
import pandas as pd                        # Read CSV files
import matplotlib.pyplot as plt            # Plot images and curves
import kagglehub                           # Download datasets from KaggleHub

import tensorflow as tf                    # Deep learning framework
from tensorflow import keras               # Keras API
from tensorflow.keras import layers        # Neural network layers
from sklearn.metrics import confusion_matrix  # Confusion matrix calculation
```

```python
# 2 Download dataset using kagglehub ---------
path = kagglehub.dataset_download("zalando-research/fashionmnist")  # Download FashionMNIST
print("Path to dataset files:", path)                               # Print local dataset folder

# 3 List all files inside the dataset folder ---------
files = sorted(glob.glob(os.path.join(path, "**", "*"), recursive=True))  # Get all files recursively
print("\nFiles found (showing first 30):")                                 # Print header
for f in files[:30]:                                                       # Show first 30 files only
    print("-", f)                                                          # Print each file path
print("Total files:", len(files))                                           # Print total count
```

```python
# --------- 4) Try to locate train/test CSV files (common in Kaggle datasets) ---------
train_csv = None                                                           # Placeholder for train CSV path
test_csv = None                                                            # Placeholder for test CSV path

for f in files:                                                            # Loop through all files
    name = os.path.basename(f).lower()                                     # Extract filename (lowercase)
    if name.endswith(".csv") and ("train" in name):                        # If it's a train CSV
        train_csv = f                                                      # Save its path
    if name.endswith(".csv") and ("test" in name):                         # If it's a test CSV
        test_csv = f                                                       # Save its path

# --------- 5) If CSV is found: read from Kaggle files. Otherwise: fallback to Keras built-in dataset ---------
use_keras_fallback = False                                                  # Flag to use built-in dataset

if (train_csv is None) or (test_csv is None):                               # If train/test CSV not found
    print("\n⚠️ No train/test CSV files found in this Kaggle download.")
    print("➡️ Falling back to keras.datasets.fashion_mnist.load_data() (easier and reliable).")
    use_keras_fallback = True                                               # Enable fallback

# --------- 6) Load data (either from CSV or from Keras) ---------
if not use_keras_fallback:                                                  # If we found Kaggle CSVs
    print("\nReading Kaggle CSV files:")
    print("Train CSV:", train_csv)
    print("Test CSV :", test_csv)

    train_df = pd.read_csv(train_csv)                                       # Read train CSV into DataFrame
    test_df  = pd.read_csv(test_csv)                                        # Read test CSV into DataFrame

    # Most FashionMNIST CSV versions have "label" column + 784 pixel columns
    y_train = train_df["label"].values                                      # Extract labels for training
    X_train = train_df.drop(columns=["label"]).values                       # Extract pixel features
    y_test  = test_df["label"].values                                       # Extract labels for testing
    X_test  = test_df.drop(columns=["label"]).values                        # Extract pixel features

    # Convert pixels to float and normalize to [0, 1]
    X_train = (X_train / 255.0).astype("float32")                           # Normalize train pixels
    X_test  = (X_test  / 255.0).astype("float32")                           # Normalize test pixels

    # Create image-shaped versions for visualization
    X_train_img = X_train.reshape(-1, 28, 28)                               # Reshape to images for plotting
    X_test_img  = X_test.reshape(-1, 28, 28)                                # Reshape to images for plotting

    # Flattened versions for MLP (Dense networks)
    X_train_flat = X_train.reshape(-1, 784)                                 # Keep as (n, 784)
    X_test_flat  = X_test.reshape(-1, 784)                                  # Keep as (n, 784)

else:                                                                       # If using Keras built-in dataset
    (X_train_img, y_train), (X_test_img, y_test) = keras.datasets.fashion_mnist.load_data()  # Load data

    # Print shapes before flattening (required by Task 0)
    print("\nBefore flatten:")
    print("X_train_img shape:", X_train_img.shape)                          # (60000, 28, 28)
    print("X_test_img shape :", X_test_img.shape)                           # (10000, 28, 28)

    # Normalize images to [0, 1]
    X_train_img = (X_train_img.astype("float32") / 255.0)                   # Normalize train images
    X_test_img  = (X_test_img.astype("float32") / 255.0)                    # Normalize test images

    # Flatten images to (784,) for Dense network
    X_train_flat = X_train_img.reshape(-1, 784)                             # Convert to (n, 784)
    X_test_flat  = X_test_img.reshape(-1, 784)                              # Convert to (n, 784)
```

```python
# --------- 7) Print shapes after flattening (Task 0 Check) ---------
print("\nAfter flatten:")
print("X_train_flat shape:", X_train_flat.shape)                            # Expected: (60000, 784)
print("X_test_flat shape :", X_test_flat.shape)                             # Expected: (10000, 784)
```

```python
# --------- 8) Visualize one random image from each class (Task 0) ---------
class_names = [
    "T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
    "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"
]                                                                            # Class labels

plt.figure(figsize=(12, 6))                                                  # Create a figure
for cls in range(10):                                                        # Loop over 10 classes
    idx = np.random.choice(np.where(y_train == cls)[0])                      # Pick a random sample index
    plt.subplot(2, 5, cls + 1)                                               # 2 rows x 5 cols
    plt.imshow(X_train_img[idx], cmap="gray")                                # Show image
    plt.title(f"{cls}: {class_names[cls]}")                                  # Show class name
    plt.axis("off")                                                          # Hide axes
plt.suptitle("One random image from each class")                             # Title
plt.tight_layout()                                                           # Layout
plt.show()                                                                   # Display figure
```

```python
# --------- 9) Create a validation split (needed for overfitting & early stopping) ---------
val_ratio = 0.1                                                              # Use 10% validation
n_val = int(len(X_train_flat) * val_ratio)                                   # Validation size

X_val = X_train_flat[:n_val]                                                 # Validation features
y_val = y_train[:n_val]                                                      # Validation labels
X_tr  = X_train_flat[n_val:]                                                 # Train features
y_tr  = y_train[n_val:]                                                      # Train labels

print("\nTrain/Val/Test shapes:")
print("Train:", X_tr.shape, "Val:", X_val.shape, "Test:", X_test_flat.shape)
```

```python
# =========================================================
# Task 1: Shallow baseline model (32 Sigmoid + Softmax)
# Optimizer: SGD | Epochs: 10
# =========================================================
baseline_model = keras.Sequential([                                           # Build a Sequential model
    layers.Input(shape=(784,)),                                               # Input layer (784 features)
    layers.Dense(32, activation="sigmoid"),                                   # Hidden layer (32 neurons, sigmoid)
    layers.Dense(10, activation="softmax")                                    # Output layer (10 classes, softmax)
])

baseline_model.compile(
    optimizer=keras.optimizers.SGD(),                                         # Use SGD optimizer
    loss=keras.losses.SparseCategoricalCrossentropy(),                        # Use sparse cross entropy
    metrics=["accuracy"]                                                      # Track accuracy
)

baseline_history = baseline_model.fit(
    X_tr, y_tr,                                                               # Training data
    validation_data=(X_val, y_val),                                           # Validation data
    epochs=10,                                                                # Train for 10 epochs
    batch_size=32,                                                            # Batch size
    verbose=1                                                                 # Show training logs
)

baseline_test_loss, baseline_test_acc = baseline_model.evaluate(
    X_test_flat, y_test, verbose=0                                            # Evaluate on test set
)
print("\nTask 1 - Baseline Test Accuracy:", baseline_test_acc)
```

```python
# =========================================================
# Task 2: Deep model (512-256-128 ReLU + Softmax)
# Optimizer: Adam | Epochs: 20
# =========================================================
deep_model = keras.Sequential([                                               # Build another Sequential model
    layers.Input(shape=(784,)),                                               # Input layer
    layers.Dense(512, activation="relu"),                                     # Hidden layer 1
    layers.Dense(256, activation="relu"),                                     # Hidden layer 2
    layers.Dense(128, activation="relu"),                                     # Hidden layer 3
    layers.Dense(10, activation="softmax")                                    # Output layer
])

deep_model.compile(
    optimizer=keras.optimizers.Adam(),                                        # Use Adam optimizer
    loss=keras.losses.SparseCategoricalCrossentropy(),                        # Loss function
    metrics=["accuracy"]                                                      # Track accuracy
)

deep_history = deep_model.fit(
    X_tr, y_tr,                                                               # Training data
    validation_data=(X_val, y_val),                                           # Validation data
    epochs=20,                                                                # Train for 20 epochs
    batch_size=32,                                                            # Batch size
    verbose=1                                                                 # Show logs
)

deep_test_loss, deep_test_acc = deep_model.evaluate(X_test_flat, y_test, verbose=0)  # Test evaluation
print("\nTask 2 - Deep Model Test Accuracy:", deep_test_acc)

print("\nComparison:")
print("Baseline Test Acc:", baseline_test_acc)
print("Deep Test Acc    :", deep_test_acc)
```

```python
# =========================================================
# Task 3: Fight overfitting (plot losses + Dropout + EarlyStopping)
# =========================================================

# ---- 3.1 Plot loss curves for the Deep Model ----
plt.figure(figsize=(7,4))                                                    # Create figure
plt.plot(deep_history.history["loss"], label="Train Loss")                   # Plot training loss
plt.plot(deep_history.history["val_loss"], label="Val Loss")                 # Plot validation loss
plt.title("Deep Model: Training vs Validation Loss")                         # Title
plt.xlabel("Epoch")                                                          # X label
plt.ylabel("Loss")                                                           # Y label
plt.legend()                                                                 # Legend
plt.show()                                                                   # Show plot

# ---- 3.2 Estimate overfitting point (first epoch where val_loss increases) ----
val_losses = deep_history.history["val_loss"]                                 # Get validation loss list
overfit_epoch = None                                                          # Placeholder
for i in range(1, len(val_losses)):                                           # Loop from second epoch
    if val_losses[i] > val_losses[i-1]:                                       # If validation loss increases
        overfit_epoch = i + 1                                                 # Save epoch number (1-indexed)
        break                                                                 # Stop at first increase
print("Estimated overfitting starts around epoch:", overfit_epoch)

# ---- 3.3 Build deep model + Dropout after each hidden layer ----
dropout_model = keras.Sequential([
    layers.Input(shape=(784,)),                                               # Input
    layers.Dense(512, activation="relu"),                                     # Hidden 1
    layers.Dropout(0.3),                                                      # Dropout (rate=0.3)
    layers.Dense(256, activation="relu"),                                     # Hidden 2
    layers.Dropout(0.3),                                                      # Dropout
    layers.Dense(128, activation="relu"),                                     # Hidden 3
    layers.Dropout(0.3),                                                      # Dropout
    layers.Dense(10, activation="softmax")                                    # Output
])

# ---- 3.4 Early stopping callback ----
early_stop = keras.callbacks.EarlyStopping(
    monitor="val_loss",                                                       # Monitor validation loss
    patience=3,                                                               # Stop after 3 epochs with no improvement
    restore_best_weights=True                                                 # Keep best weights
)

dropout_model.compile(
    optimizer=keras.optimizers.Adam(),                                        # Adam optimizer
    loss=keras.losses.SparseCategoricalCrossentropy(),                        # Loss
    metrics=["accuracy"]                                                      # Accuracy
)

dropout_history = dropout_model.fit(
    X_tr, y_tr,                                                               # Train
    validation_data=(X_val, y_val),                                           # Validation
    epochs=40,                                                                # Max epochs (EarlyStopping will end earlier)
    batch_size=32,                                                            # Batch size
    callbacks=[early_stop],                                                   # Callback list
    verbose=1                                                                 # Show logs
)

drop_test_loss, drop_test_acc = dropout_model.evaluate(X_test_flat, y_test, verbose=0)  # Test evaluation
print("\nTask 3 - Dropout+EarlyStopping Test Accuracy:", drop_test_acc)

# ---- 3.5 Plot new loss curves (regularized model) ----
plt.figure(figsize=(7,4))
plt.plot(dropout_history.history["loss"], label="Train Loss")
plt.plot(dropout_history.history["val_loss"], label="Val Loss")
plt.title("Dropout+EarlyStopping: Training vs Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()

# ---- 3.6 Check train vs validation accuracy gap (last recorded epoch) ----
train_acc_last = dropout_history.history["accuracy"][-1]                      # Last train accuracy
val_acc_last = dropout_history.history["val_accuracy"][-1]                    # Last val accuracy
print("Train acc (last):", train_acc_last)
print("Val acc (last)  :", val_acc_last)
print("Gap (train - val):", train_acc_last - val_acc_last)
```

```python
# =========================================================
# Task 4: Forensic analysis (Confusion Matrix + Heatmap + Misclassified images)
# =========================================================

# ---- 4.1 Predict labels on test set ----
y_prob = dropout_model.predict(X_test_flat, verbose=0)                         # Get probabilities
y_pred = np.argmax(y_prob, axis=1)                                             # Convert probs to predicted class

# ---- 4.2 Confusion matrix ----
cm = confusion_matrix(y_test, y_pred)                                          # Calculate confusion matrix
print("\nConfusion Matrix:\n", cm)

# ---- 4.3 Heatmap using matplotlib (no seaborn needed) ----
plt.figure(figsize=(8,6))
plt.imshow(cm, interpolation="nearest", cmap="Blues")                          # Show matrix as image
plt.title("Confusion Matrix (Best Model)")
plt.colorbar()
ticks = np.arange(10)
plt.xticks(ticks, class_names, rotation=45, ha="right")
plt.yticks(ticks, class_names)
plt.xlabel("Predicted")
plt.ylabel("Actual")

# Add numbers to each cell
thresh = cm.max() * 0.6
for i in range(10):
    for j in range(10):
        plt.text(j, i, cm[i, j],
                 ha="center", va="center",
                 color="white" if cm[i, j] > thresh else "black",
                 fontsize=9)
plt.tight_layout()
plt.show()

# ---- 4.4 Find the most confused pair (largest off-diagonal) ----
cm_off = cm.copy()                                                             # Copy matrix
np.fill_diagonal(cm_off, 0)                                                    # Ignore correct predictions
i, j = np.unravel_index(np.argmax(cm_off), cm_off.shape)                       # Index of largest confusion
print(f"Most confused pair: Actual '{class_names[i]}' predicted as '{class_names[j]}' count={cm_off[i, j]}")

# Check Sneaker (7) vs Ankle boot (9)
print("Sneaker (7) predicted as Ankle boot (9):", cm[7, 9])
print("Ankle boot (9) predicted as Sneaker (7):", cm[9, 7])

# ---- 4.5 Plot 3 misclassified images ----
mis_idx = np.where(y_pred != y_test)[0]                                        # Indices where prediction is wrong
print("Total misclassified:", len(mis_idx))

# Select 3 random misclassified samples (if enough exist)
pick = np.random.choice(mis_idx, size=3, replace=False) if len(mis_idx) >= 3 else mis_idx

plt.figure(figsize=(10,4))
for k, idx in enumerate(pick):
    plt.subplot(1, 3, k+1)
    plt.imshow(X_test_img[idx], cmap="gray")
    plt.axis("off")
    plt.title(f"Predicted: {class_names[y_pred[idx]]}\nActual: {class_names[y_test[idx]]}")
plt.suptitle("Misclassified Examples")
plt.tight_layout()
plt.show()
```

```python
# =========================================================
# Bonus (Optional): Batch size experiment (32 vs 1024)
# =========================================================

def build_dropout_model():
    """Build the same regularized model again (for fair batch size comparison)."""
    m = keras.Sequential([
        layers.Input(shape=(784,)),
        layers.Dense(512, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(256, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(128, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(10, activation="softmax")
    ])
    m.compile(
        optimizer=keras.optimizers.Adam(),
        loss=keras.losses.SparseCategoricalCrossentropy(),
        metrics=["accuracy"]
    )
    return m

def train_with_batch(batch_size, epochs=15):
    """Train model with a specific batch size and return time + test accuracy."""
    model = build_dropout_model()
    es = keras.callbacks.EarlyStopping(monitor="val_loss", patience=3, restore_best_weights=True)
    start = time.time()
    model.fit(
        X_tr, y_tr,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[es],
        verbose=0
    )
    elapsed = time.time() - start
    _, test_acc = model.evaluate(X_test_flat, y_test, verbose=0)
    return elapsed, test_acc

t32, acc32 = train_with_batch(32)
t1024, acc1024 = train_with_batch(1024)

print("\nBONUS RESULTS:")
print(f"Batch=32   -> time={t32:.2f}s, test_acc={acc32:.4f}")
print(f"Batch=1024 -> time={t1024:.2f}s, test_acc={acc1024:.4f}")
```
