```python
import tensorflow as tf

num_classes = 10 

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1000, 1000)),   # صورة 1000x1000
    tf.keras.layers.Flatten(),                   # -> 1,000,000 features
    tf.keras.layers.Dense(1000, activation="relu"),  # Hidden layer 1
    tf.keras.layers.Dense(num_classes, activation="softmax")  # Output
])

model.summary()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

# =========================
# 1. Generate Synthetic Data
# =========================

# Keep this VERY small
num_samples = 5

# Random 1000x1000 images
X = np.random.rand(num_samples, 1000, 1000).astype(np.float32)

# Binary labels
y = np.zeros((num_samples, 1), dtype=np.float32)

# Add a bright center square for class = 1
for i in range(num_samples):
    if i % 2 == 0:
        X[i, 450:550, 450:550] += 2.0
        y[i] = 1.0

print("X shape:", X.shape)
print("y shape:", y.shape)

# =========================
# 2. Define the Model
# =========================

model = Sequential([
    Flatten(input_shape=(1000, 1000)),   # 1,000,000 inputs
    Dense(1000, activation='relu'),      # Hidden layer 1
    Dense(1000, activation='relu'),      # Hidden layer 2
    Dense(1000, activation='relu'),      # Hidden layer 3
    Dense(1, activation='sigmoid')       # Binary output
])

model.summary()
```

```python

# =========================
# 3. Compile the Model
# =========================

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# =========================
# 4. Train the Model
# =========================

model.fit(
    X,
    y,
    epochs=2,
    batch_size=1
)

# =========================
# 5. Test Prediction
# =========================

pred = model.predict(X)
print("Predictions:\n", pred)
```
