```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer
```

```python
df_raw = pd.read_csv('merged_store_sales.csv')
df = df_raw.copy()
```

```python
df
```

```python
def fill_numeric(df):
    df = df.copy()
    df["Qty"] = df["Qty"].fillna(0).astype(int)
    df["Unit_Price"] = df["Unit_Price"].fillna(0.0).astype(float)
    return df

def fill_customers(df):
    df = df.copy()
    df["CustomerID"] = df["CustomerID"].fillna("GUEST")
    return df

def convert_date(df):
    df = df.copy()
    df["SaleDate"] = pd.to_datetime(df["SaleDate"])
    return df

def calc_total(df):
    df = df.copy()
  
    df["Total_Price"] = df["Qty"] * df["Unit_Price"]
    return df


def convert_currency(df):
    df = df.copy()
    rates = {"OMR": 1, "USD": 0.385, "EUR": 0.42}

   
    df["Total_Price"] = df.apply(
        lambda r: r["Total_Price"] * rates.get(str(r["CurrencyType"]).upper(), 1),
        axis=1
    )
    return df

def fill_currency(df):
    df = df.copy()
    mode_value = df["CurrencyType"].mode()[0]   
    df["CurrencyType"] = df["CurrencyType"].fillna(mode_value)
    return df
```

```python
pipeline = Pipeline([
    ("fill_numeric", FunctionTransformer(fill_numeric)),
    ("fill_customers", FunctionTransformer(fill_customers)),
    ("fill_currency", FunctionTransformer(fill_currency)),  # ← هنا نضفناها
    ("convert_date", FunctionTransformer(convert_date)),
    ("calc_total", FunctionTransformer(calc_total)),
    ("convert_currency", FunctionTransformer(convert_currency)),
])
```

```python
df_transformed = pipeline.fit_transform(df_raw)
print(df_transformed.head())
```
