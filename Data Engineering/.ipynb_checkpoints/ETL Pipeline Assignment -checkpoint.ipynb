```python
import numpy as np
import pandas as pd
import pymysql
from sklearn.preprocessing import FunctionTransformer, LabelEncoder, StandardScaler
from sklearn.pipeline import Pipeline
```

```python
df=pd.read_csv("train.csv")
df
```

```python
#● Display the first few rows to understand the dataset. 
print(df.head())
```

```python
df.columns
```

```python
# Select only the required columns
def select_columns(df):
    # Make a copy to avoid modifying the original DataFrame
    df = df.copy()
    cols = ["PassengerId", "Survived", "Pclass", "Sex", "Age", "Fare", "Embarked"]
    df = df[cols]
    return df

# Handle missing values for Age (median) and Embarked (most frequent)
def handle_missing(df):
    # Make a copy to avoid modifying the original DataFrame
    df = df.copy()
    # Fill missing Age with the median value
    df["Age"] = df["Age"].fillna(df["Age"].median())
    # Fill missing Embarked with the most frequent value (mode)
    df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
    return df

# Prepare encoders and scaler as global objects
gender_encoder = LabelEncoder()
embarked_encoder = LabelEncoder()
scaler = StandardScaler()

# Encode categorical columns: Sex (Gender) and Embarked
def encode_categorical(df):
    # Make a copy to avoid modifying the original DataFrame
    df = df.copy()
    # Apply label encoding to the Sex column
    df["Sex"] = gender_encoder.fit_transform(df["Sex"])
    # Apply label encoding to the Embarked column
    df["Embarked"] = embarked_encoder.fit_transform(df["Embarked"])
    return df

# Scale numeric columns: Age and Fare using StandardScaler
def scale_numeric(df):
    # Make a copy to avoid modifying the original DataFrame
    df = df.copy()
    # Apply StandardScaler to Age and Fare columns
    df[["Age", "Fare"]] = scaler.fit_transform(df[["Age", "Fare"]])
    return df

# Build the transformation pipeline
pipeline = Pipeline([
    ("select_columns", FunctionTransformer(select_columns)),
    ("handle_missing", FunctionTransformer(handle_missing)),
    ("encode_categorical", FunctionTransformer(encode_categorical)),
    ("scale_numeric", FunctionTransformer(scale_numeric)),
])

# Apply the pipeline to the original Titanic DataFrame df
df_transformed = pipeline.fit_transform(df)

# Check for any remaining missing values
print(df_transformed.isna().sum())

# Display first rows of the final cleaned DataFrame
print(df_transformed.head())
```

B. Create a table named:  titanic_clean 

```python
# Create table that matches df_transformed columns
cursor.execute("""
DROP TABLE IF EXISTS titanic_clean;
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS titanic_clean (
    PassengerId INT,
    Survived INT,
    Pclass INT,
    Sex INT,          -- encoded gender (e.g., 0, 1)
    num__Age FLOAT,   -- scaled age
    num__Fare FLOAT,  -- scaled fare
    Embarked INT      -- encoded embarked (e.g., 0,1,2)
);
""")

print("Table 'titanic_clean' created successfully!")
```

```python
# C. Load the cleaned data into MySQL

# Prepare the INSERT statement
insert_sql = """
INSERT INTO titanic_clean (
    PassengerId, Survived, Pclass,
    Sex, num__Age, num__Fare,
    Embarked
)
VALUES (%s,%s,%s,%s,%s,%s,%s);
"""

# Loop over each row in the cleaned DataFrame
for _, row in df_transformed.iterrows():
    cursor.execute(insert_sql, (
        int(row["PassengerId"]),   # Passenger ID
        int(row["Survived"]),      # Survival status
        int(row["Pclass"]),        # Passenger class
        int(row["Sex"]),           # Encoded gender
        float(row["num__Age"]),    # Scaled age
        float(row["num__Fare"]),   # Scaled fare
        int(row["Embarked"])       # Encoded embarked port
    ))

# Commit all inserts to the database
connection.commit()

print("All cleaned rows inserted successfully into 'titanic_clean'!")
```
