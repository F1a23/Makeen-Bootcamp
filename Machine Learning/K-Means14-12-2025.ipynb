```python
marks = [35, 40, 42, 60, 65, 70]
c1 = 35.0
c2 = 70.0
while True:
    cluster1 = []
    cluster2 = []
    for x in marks:
        if abs(x - c1) <= abs(x - c2):
            cluster1.append(x)
        else:
            cluster2.append(x)
    new_c1 = sum(cluster1) / len(cluster1)
    new_c2 = sum(cluster2) / len(cluster2)
    
    print("All data in cluster1 are:", cluster1, "And the final centroid1 is:", c1)
    print("All data in cluster2 are:", cluster2, "And the final centroid2 is:", c2)

    if new_c1 == c1 and new_c2 == c2:
        break
    c1, c2 = new_c1, new_c2
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# 1) Load data
df = pd.read_csv("03_Clustering_Marketing.csv")

# 2) Only 10 columns
cols_10 = [
    "gradyear", "gender", "age", "NumberOffriends",
    "basketball", "football", "soccer", "music",
    "shopping", "drugs"
]

X = df[cols_10].copy()

# 3) Convert everything to numeric 
for col in X.columns:
    X[col] = pd.to_numeric(X[col], errors="coerce")

# 4) Fill Nulls (no deleting)
X = X.fillna(0)

# 5) Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 6) Find best K
best_k = 0
best_score = -1

print("Silhouette Scores:")
for k in range(2, 11):
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = model.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    print(f"K = {k} → Score = {score:.4f}")

    if score > best_score:
        best_score = score
        best_k = k

print("\nBest K =", best_k)
print("Best Silhouette Score =", round(best_score, 4))

# 7) Final model
final_kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
df["Cluster"] = final_kmeans.fit_predict(X_scaled)

print("\nCluster counts:")
print(df["Cluster"].value_counts())
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

df = pd.read_csv("03_Clustering_Marketing.csv")

cols_10 = ["gradyear","gender","age","NumberOffriends",
           "basketball","football","soccer","music","shopping","drugs"]

X = df[cols_10].apply(pd.to_numeric, errors="coerce").fillna(0)

X_scaled = StandardScaler().fit_transform(X)

best_k, best_score = 0, -1

for k in range(2, 11):
    labels = KMeans(n_clusters=k, random_state=42, n_init=10).fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    print(k, score)

    if score > best_score:
        best_k, best_score = k, score

print("Best K =", best_k, "| Best score =", best_score)

df["Cluster"] = KMeans(n_clusters=best_k, random_state=42, n_init=10).fit_predict(X_scaled)
print(df["Cluster"].value_counts())
```
