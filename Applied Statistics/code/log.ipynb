```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#non-normal 
data = {'income': [2000, 3000, 5000, 10000, 20000, 40000, 80000, 150000]}
df = pd.DataFrame(data)

# Apply log transformation
df['log_income'] = np.log(df['income'])

# Plot the distribution before and after transformation
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(df['income'], bins=10, edgecolor='black')
plt.title('Original Income Distribution')

plt.subplot(1,2,2)
plt.hist(df['log_income'], bins=10, edgecolor='black')
plt.title('Log-Transformed Income Distribution')

plt.tight_layout()
plt.show()
```
