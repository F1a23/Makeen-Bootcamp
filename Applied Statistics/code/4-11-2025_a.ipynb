```python
salaries = [500, 520, 530, 540, 550, 9000] 

#1. Compute mean, median, and mode.
mean=round(sum(salaries)/len(salaries),2)

salaries.sort()
n = len(salaries)
if n % 2 == 0:
    median = round((salaries[n//2 - 1] + salaries[n//2]) / 2, 2)
else:
    median = round(salaries[n//2], 2)

mode = round(max(salaries, key=salaries.count),2)

print("The Mean is :",mean)
print("The Median is :",median)
print("The Mode is :",mode) 


print("-"*100)

#2. Remove the outlier (9000) and compute again. 
#IQR=Q1-Q3
Q1 = np.percentile(salaries, 25)
Q3 = np.percentile(salaries, 75)
IQR = Q3 - Q1
#print(IQR)
low  = Q1 - 1.5 * IQR
high = Q3 + 1.5 * IQR

#print(low)
#print(high)

outliers = []
clean_data = []

for x in salaries:
    if x < low or x > high:
        outliers.append(x)
    else:
        clean_data.append(x)

print("Outliers:", outliers)
print("Clean data:", clean_data)
print("---")

mean2 = round(sum(clean_data) / len(clean_data), 2)

clean_data.sort()
n = len(clean_data)
if n % 2 == 0:
    median2 = round((clean_data[n//2 - 1] + clean_data[n//2]) / 2, 2)
else:
    median2 = round(clean_data[n//2], 2)

mode2 = round(max(clean_data, key=clean_data.count), 2)

print("The Mean after removing the outlier:", mean2)
print("The Median after removing the outlier:", median2)
print("The Mode after removing the outlier:", mode2)
print("-"*100)
# 3. Compare the results.
print("\nComparison:")
print(f"Mean changed from {mean} → {mean2}")
print(f"Median changed from {median} → {median2}")
print(f"Mode stayed the same at {mode}")
```
