```python
import numpy as np 
import pandas as pd 
```

Part 1 – Measures of Central Tendency 

Exercise 1: Mean, Median, and Mode 

```python
gpas =[2.7, 3.0, 3.5, 3.8, 3.0, 3.2, 3.4, 3.0, 3.1, 3.9] 
df = pd.DataFrame(gpas, columns=['GPA'])

mean = round(df['GPA'].mean(), 2)
median = round(df['GPA'].median(), 2)
mode = round(df['GPA'].mode()[0], 2)
#df
print("The Mean is:", mean)
print("The Median is:", median)
print("The Mode is:", mode)
```

```python
print("""
1.What is the typical GPA of this group? 
    3.26 — shows generally good performance.
        
2. Which measure (mean, median, or mode) best represents the data — and why? 
    Median (3.15) — it best represents the middle of the data and isn’t affected by outliers.
        
3. If one student’s GPA changes to 1.5, which measure will be most affected? 
    Most affected if GPA = 1.5:   Mean — it changes the most because it’s sensitive to extreme values.
""")
```

Part 2 – Measures of Variability 

Exercise 2: Effect of Outliers 

```python
salaries = [500, 520, 530, 540, 550, 9000] 
df1 = pd.DataFrame(salaries, columns=['salaries'])

#1. Compute mean, median, and mode.
mean =df1['salaries'].mean()
median =df1['salaries'].median()
mode =df1['salaries'].mode()[0]
print("The Mean is:", mean)
print("The Median is:", median)
print("The Mode is:", mode)

print("\n")
print("*"*100)
#2. Remove the outlier (9000) and compute again. 
df_no_outlier = df1[df1['salaries'] != 9000]
# Compute again
mean2 = df_no_outlier['salaries'].mean()
median2 = df_no_outlier['salaries'].median()
mode2 = df_no_outlier['salaries'].mode()[0]
print("\nAfter removing outlier:")
print("Mean:", mean2)
print("Median:", median2)
print("Mode:", mode2)
print("\n")
print("*"*100)

# 3. Compare the results.
print("\nComparison:")
print(f"Mean changed from {mean} → {mean2}")
print(f"Median changed from {median} → {median2}")
print(f"Mode stayed the same at {mode}")
```

Part 2 – Measures of Variability 

Exercise 3: Range, Variance, and Standard Deviation 

```python
ages = [18, 19, 20, 20, 21, 22, 22, 23, 24, 30]
df2 = pd.DataFrame(ages, columns=['ages'])

# 1. Calculate range, variance, and standard deviation
age_range = df2['ages'].max() - df2['ages'].min()
variance = round(df2['ages'].var(), 2)
std_dev = round(df2['ages'].std(), 2)

# 2. Display results
print("Range:", age_range)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
group_A = [65, 67, 68, 69, 70, 72, 74, 75]
group_B = [60, 63, 67, 70, 75, 78, 80, 85]

df_A = pd.DataFrame(group_A, columns=['Group A'])
df_B = pd.DataFrame(group_B, columns=['Group B'])

# 1. Mean and Standard Deviation (rounded to 2 decimals)
mean_A = round(df_A['Group A'].mean(), 2)
std_A = round(df_A['Group A'].std(), 2)

mean_B = round(df_B['Group B'].mean(), 2)
std_B = round(df_B['Group B'].std(), 2)
#2. Compare the spread. 
print("Group A → Mean:", mean_A, "| Std Dev:", std_A)
print("Group B → Mean:", mean_B, "| Std Dev:", std_B)
print("\n")
print("""Comparison:
Group A has scores close to the mean (smaller spread),
while Group B’s scores vary more widely (larger standard deviation).
""")
```

```python
print("""
Interpretation Questions:

• Which group is more consistent?
→ Group A is more consistent because it has a smaller standard deviation (3.46).

• If both groups have similar means, why does standard deviation still matter?
→ Standard deviation shows how spread out the scores are — Group B’s higher SD means more variation.

• What can you infer about performance differences between groups?
→ Both perform similarly on average, but Group B has wider score differences among members.
""")
```
