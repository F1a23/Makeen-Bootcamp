```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Gradas=[40,55,60,65,70,80,85,90,95]
k = 2
mean = np.mean(Gradas)
print("the Mean is:", mean)

s = np.std(Gradas, ddof=1)
print("SD is:", s)

cv = round((s / mean) * 100, 0) #cv is the persentage of version from the mean
print("CV is:", cv, "%")

at_least = (1 - (1 / k**2)) * 100
print("at_least:", at_least)

lower = mean - k * s
upper = mean + k * s


count_number = np.sum((Gradas >= lower) & (Gradas <= upper))
total = len(Gradas)
print("the total number is:", total)

percent = round((count_number / total) * 100)
print("the percent is:", percent)
```

```python
scores = [50, 60, 70, 80, 90]
df = pd.DataFrame(scores, columns=['Score'])

mean = df['Score'].mean()
std = round(df['Score'].std(),1)

df['Z-Score'] = (df['Score'] - mean) // std

print("The Mean is:", mean)
print("The Standard Deviation is:", std)
print(df)

plt.hist(df['Z-Score'], bins=5, edgecolor='black')
plt.title("Histogram of Z-Scores")
plt.xlabel("Z-Score")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
scores = [50,60,60,70,70,70,80,80,90]
mean = np.mean(scores)
print('the mean:',mean)
std = np.std(scores)
print('the std:',std)
z= [(x-mean) / std for x in scores]

plt.hist(z, bins=5, edgecolor="black")
plt.title("Scores")
plt.xlabel("Std Scores (Z)")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm  #    

# 
scores = [50, 60, 60, 70, 70, 70, 80, 80, 90]

#    
mean = np.mean(scores)
print('The mean:', mean)
std = np.std(scores)
print('The std:', std)

#    Z
z = [(x - mean) / std for x in scores]

#  
plt.hist(z, bins=5, density=True, edgecolor="black", color='skyblue', alpha=0.7)

#    
x = np.linspace(min(z)-1, max(z)+1, 100)
y = norm.pdf(x, 0, 1)  #     0   1

#     
plt.plot(x, y, color='red', linewidth=2)

# 
plt.title("Histogram of Z-Scores with Normal Distribution Curve")
plt.xlabel("Z-Score")
plt.ylabel("Density")
plt.grid(True, linestyle='--', alpha=0.7)

#  
plt.show()
```
